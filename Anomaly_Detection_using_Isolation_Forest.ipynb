```python
# Required Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
# Generate synthetic data
np.random.seed(42)
normal_data = np.random.randn(1000, 2)* 2  # Normal data points
anomaly_data = np.random.randn(50, 2)* 10  # Normal data points

# Combine normal and anomaly data
data=  np.vstack((normal_data, anomaly_data))

# Fit isolationForest Model
model = IsolationForest(contamination=0.05)

# Contamination parameter sets expected proportion of anomalies
model.fit(data)
```

```python
# predict anomalies
predictions = model.predict(data)

# Plot the data and detected anomalies
plt.scatter(data[:, 0], data[:, 1], c = predictions, cmap=plt.cm.Paired)
plt.title("Anomaly Detection Using Isolation Forest")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

# Implementing Isolation Forest on Iris Dataset

```python
# Importing required dataset
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
# Get the data
iris = load_iris()
X = iris.data
print(X)
```

```python
# Standardize features to have mean=0 and variance=1
sc = StandardScaler()
X = sc.fit_transform(X)

# Define the model
clf = IsolationForest(contamination=0.1)

# Fit the model
clf.fit(X)

# Predict the anomalies in the data
pred = clf.predict(X)

# Print the anomaly prediction {-1:anomaly, 1:normal}
print(pred)

plt.scatter(X[:, 0], X[:, 1], c = pred, cmap=plt.cm.Paired)
plt.title("Anomaly Detection Using Isolation Forest")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
